```python
import numpy as np
import matplotlib.pyplot as plt

from making_synthetic_blobs import Generate_synthetic_data

from manual_matching import match_synthetic_data

```

# dependece of sucess rate on the blobs' error:

```python
repetitions = 10
Np = 150

s = 2*20.0
r_avg = (s**3/Np)**(1.0/3.0)
print('average interparticle distance: ',r_avg)


error_lst = np.linspace(0, 5.0, num=20)
Np_list = [50, 250, 500]

voxel_size = 2.0
max_err = 10.0


success_list = []
wrong_list = []
position_error_list = []

for Np in Np_list:
    success_list.append([])
    wrong_list.append([])
    position_error_list.append([])
    for err in error_lst:
        tmp_sl = []
        tmp_wl = []
        tmp_err = []

        for i in range(repetitions):

            Generate_synthetic_data(Np, err, generate_images=False)
            success, wrong, avg_error, runtime = match_synthetic_data(voxel_size, max_err, save_name = 'particles')

            if success+wrong>Np:
                duplicates = success+wrong - Np
            else:
                duplicates = 0

            tmp_sl.append(success/Np)
            tmp_wl.append(wrong/Np)
            tmp_err.append(avg_error)

        success_list[-1].append(np.mean(tmp_sl))
        wrong_list[-1].append(np.mean(tmp_wl))
        position_error_list[-1].append(np.mean(tmp_err))
        
```

```python
fig, ax = plt.subplots(2,2, sharex=True)
fig.set_size_inches(10,6)


lbls = ['Np=50', 'Np=250', 'Np=500']

for i in range(len(Np_list)):
    ax[0][0].plot(error_lst, success_list[i], 'o-', ms=2, lw=.5, label=lbls[i])
    ax[1][0].plot(error_lst, wrong_list[i], 'o-', ms=2, lw=.5)
    tot = np.array(wrong_list[i]) + np.array(success_list[i])
    ax[0][1].plot(error_lst, tot, 'o-', ms=2, lw=.5)
    ax[1][1].semilogy(error_lst, position_error_list[i], 'o-', ms=2, lw=.5)

    
ax[0][0].set_ylim(0, 1.1)
ax[0][0].set_ylabel('correct/Np')
ax[0][0].grid()

ax[1][0].set_ylim(0, 1.1)
ax[1][0].set_ylabel('wrong/Np')
ax[1][0].grid()

ax[0][1].set_ylim(0.0, 1.1)
ax[0][1].set_ylabel('matched/Np')
ax[0][1].grid()

ax[1][1].set_ylim(bottom=0.01, top= r_avg/2)
ax[1][1].set_ylabel('triangulation error [mm]')
ax[1][1].grid()

ax[1][0].set_xlabel('pixel error')
ax[1][1].set_xlabel('pixel error')

ax[0][0].legend()

plt.tight_layout()
```

```python
fig.savefig('pixel_error_effect.pdf')
```

# Voxel size effect:

```python
repetitions = 5
Np = 100

s = 2*20.0
r_avg = (s**3/150)**(1.0/3.0)
print('average interparticle distance: ',r_avg)


err = 2.0

voxel_size_lst = [0.1, 0.2, 0.4, 0.8, 1.5, 3, 5, 8, 12]
max_err_lst = [0.2, 0.5, 1.0, 2.0, 4.0]


success_list = []
wrong_list = []
position_error_list = []

for max_err in max_err_lst:
    success_list.append([])
    wrong_list.append([])
    position_error_list.append([])
    for vs in voxel_size_lst[::-1]:
        tmp_sl = []
        tmp_wl = []
        tmp_err = []

        for i in range(repetitions):

            Generate_synthetic_data(Np, err, generate_images=False)
            success, wrong, avg_error, runtime = match_synthetic_data(vs, max_err, save_name = 'particles')
            tmp_sl.append(success/Np)
            tmp_wl.append(wrong/Np)
            tmp_err.append(avg_error)

        success_list[-1].append(np.mean(tmp_sl))
        wrong_list[-1].append(np.mean(tmp_wl))
        position_error_list[-1].append(np.mean(tmp_err))
        
```

```python
fig, ax = plt.subplots(2,2, sharex=True)
fig.set_size_inches(10,6)

lbls = ['max_err=0.2', 'max_err=0.5', 'max_err=1.0', 'max_err=2.0', 'max_err=4.0']
for i in range(len(max_err_lst)):
    ax[0][0].plot(voxel_size_lst, success_list[i], 'o-', ms=2, lw=.5)
    ax[1][0].plot(voxel_size_lst, wrong_list[i], 'o-', ms=2, lw=.5)
    tot = np.array(wrong_list[i]) + np.array(success_list[i])
    ax[0][1].plot(voxel_size_lst, tot, 'o-', ms=2, lw=.5, label=lbls[i])
    ax[1][1].semilogy(voxel_size_lst, position_error_list[i], 'o-', ms=2, lw=.5)

    
ax[0][0].set_ylim(bottom = -0.01)
ax[0][0].set_ylabel('correct/Np')
ax[0][0].grid()

ax[1][0].set_ylim(-0.01, 1.1)
ax[1][0].set_ylabel('wrong/Np')
ax[1][0].grid()

ax[0][1].set_ylim(-0.01, max([1.1, max(tot)+0.1]))
ax[0][1].set_ylabel('matched/Np')
ax[0][1].grid()

ax[1][1].set_ylim(bottom=0.01, top= r_avg/2)
ax[1][1].set_ylabel('triangulation error [mm]')
ax[1][1].grid()

ax[1][0].set_xlabel('voxel size')
ax[1][1].set_xlabel('voxel size')

ax[0][1].legend(ncol=2)

fig.suptitle('voxel size effect, tested with pixel error=2.0;  Np=100')

plt.tight_layout()
```

# maximum matching error effect:

```python
repetitions = 5
Np = 100

s = 2*20.0
r_avg = (s**3/150)**(1.0/3.0)
print('average interparticle distance: ',r_avg)




voxel_size = 2.0
max_err_lst = np.append(np.linspace(0.001, 0.1, num=10), np.linspace(0.1, 1.0, num=7))


success_list = []
wrong_list = []
position_error_list = []

for max_err in max_err_lst:
    tmp_sl = []
    tmp_wl = []
    tmp_err = []

    for i in range(repetitions):

        Generate_synthetic_data(Np, err, generate_images=False)
        success, wrong, avg_error, runtime = match_synthetic_data(voxel_size, max_err, save_name = 'particles')

        if success+wrong>Np:
            duplicates = success+wrong - Np
        else:
            duplicates = 0

        tmp_sl.append(success/Np)
        tmp_wl.append(wrong/Np)
        tmp_err.append(avg_error)

    success_list.append(np.mean(tmp_sl))
    wrong_list.append(np.mean(tmp_wl))
    position_error_list.append(np.mean(tmp_err))
        
```

```python
fig, ax = plt.subplots(2,2, sharex=True)
fig.set_size_inches(10,6)



ax[0][0].plot(max_err_lst, success_list, 'o-', ms=2, lw=.5)
ax[1][0].plot(max_err_lst, wrong_list, 'o-', ms=2, lw=.5)
tot = np.array(wrong_list) + np.array(success_list)
ax[0][1].plot(max_err_lst, tot, 'o-', ms=2, lw=.5)
ax[1][1].semilogy(max_err_lst, position_error_list, 'o-', ms=2, lw=.5)

    
ax[0][0].set_ylim(-0.01, 1.1)
ax[0][0].set_ylabel('correct/Np')
ax[0][0].grid()

ax[1][0].set_ylim(-0.01, 1.1)
ax[1][0].set_ylabel('wrong/Np')
ax[1][0].grid()

ax[0][1].set_ylim(-0.01, 1.1)
ax[0][1].set_ylabel('matched/Np')
ax[0][1].grid()

ax[1][1].set_ylim(bottom=0.01, top= r_avg/2)
ax[1][1].set_ylabel('triangulation error [mm]')
ax[1][1].grid()

ax[1][0].set_xlabel('max_err size')
ax[1][1].set_xlabel('max_err size')

fig.suptitle('maximum allowed matching error effect: pixel error=2.0;  Np=100')

plt.tight_layout()
```
